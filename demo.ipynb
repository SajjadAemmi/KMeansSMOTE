```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from imblearn.datasets import fetch_datasets
from kmeans_smote import KMeansSMOTE
```

```python
datasets = fetch_datasets()
X, y = datasets['oil']['data'], datasets['oil']['target']
```

```python
for label, count in zip(*np.unique(y, return_counts=True)):
    print('Class {} has {} instances'.format(label, count))
```

```python
cmap = ['red', 'blue']
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=ListedColormap(cmap), marker='.', alpha=0.5)
plt.show()
```

```python
kmeans_smote = KMeansSMOTE(
    kmeans_args={
        'n_clusters': 100,
        'n_init': 3
    },
    smote_args={
        'k_neighbors': 10
    }
)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)
```

```python
for label, count in zip(*np.unique(y_resampled, return_counts=True)):
    print('Class {} has {} instances after oversampling'.format(label, count))
```

```python
cmap = ['red', 'blue']
plt.scatter(X_resampled[:, 0], X_resampled[:, 1], c=y_resampled, cmap=ListedColormap(cmap), marker='.', alpha=0.5)
plt.show()
```
